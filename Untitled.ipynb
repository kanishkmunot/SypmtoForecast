```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as ply
import seaborn as sns
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
# Create a list of symptom columns
symptom_columns = ['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4', 'Symptom_5', 'Symptom_6', 'Symptom_7', 'Symptom_8', 'Symptom_9', 'Symptom_10', 'Symptom_11', 'Symptom_12', 'Symptom_13', 'Symptom_14', 'Symptom_15', 'Symptom_16', 'Symptom_17']

# Perform one-hot encoding for symptom columns
df_encoded = pd.get_dummies(df, columns=symptom_columns)

# Display the first few rows of the encoded DataFrame
df_encoded.head()
```

```python
# Define the columns to replace values in (excluding 'Disease' column)
columns_to_replace = df_encoded.columns.drop('Disease')

# Replace 'True' with 1 and 'False' with 0 in selected columns
df_encoded[columns_to_replace] = df_encoded[columns_to_replace].replace({True: 1, False: 0})

# Display the first few rows of the encoded DataFrame
df_encoded.head()
```

```python
# Find duplicate rows
duplicate_rows = df_encoded[df_encoded.duplicated()]

# Display the duplicate rows
print("Duplicate Rows:")
print(duplicate_rows)

# Remove duplicate rows
df_encoded = df_encoded.drop_duplicates()

# Display the shape of the DataFrame after removing duplicates
df_encoded.shape
```

```python
from sklearn.model_selection import train_test_split

# Separate the target variable (Disease) from the features
X = df_encoded.drop(columns=['Disease'])
y = df_encoded['Disease']

# Split the data into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Display the shapes of the subsets
print("Shapes of subsets:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict, train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Split the data into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Initialize the Random Forest classifier
rf_model = RandomForestClassifier(random_state=42)

# Perform k-fold cross-validation and get predictions
y_pred_cv = cross_val_predict(rf_model, X_train, y_train, cv=5)

# Train the model on the entire training data
rf_model.fit(X_train, y_train)

# Predict on the testing data
y_pred = rf_model.predict(X_test)

# Calculate various performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Display performance metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)

# Generate and display the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```
